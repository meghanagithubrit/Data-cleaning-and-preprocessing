1. Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
plt.style.use("ggplot")
sns.set_theme(style="whitegrid")
```

```python
url = "/content/archive.zip"
df = pd.read_csv(url)
```

Step 1: Summary **Statistics**

```python
print("===== Dataset Info =====")
print(df.info())
print("\n===== First 5 Rows =====")
print(df.head())
print("\n===== Summary Statistics =====")
print(df.describe(include="all"))
```

```python
print("\n===== Missing Values =====")
print(df.isnull().sum())
```

Step 2: Histograms & **Boxplots**

```python
numeric_cols = df.select_dtypes(include=["int64","float64"]).columns
```

```python
df[numeric_cols].hist(figsize=(12,8), bins=20, color="skyblue")
plt.suptitle("Histograms of Numeric Features")
plt.show()
```

```python
for col in numeric_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(y=df[col], color="lightgreen")
    plt.title(f"Boxplot of {col}")
    plt.show()
```

**bold text**

Step 3: Correlation & Pairplot

```python
plt.figure(figsize=(8,6))
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.pairplot(df[numeric_cols].dropna().sample(200), diag_kind="kde")
plt.show()
```

Step 4: Trends / **Anomalies**

```python
if "Survived" in df.columns and "Pclass" in df.columns:
    plt.figure(figsize=(6,4))
    sns.barplot(x="Pclass", y="Survived", data=df, ci=None, palette="Set2")
    plt.title("Survival Rate by Passenger Class")
    plt.show()
```

Step 5: Interactive Plotly **Visualization**

```python
if "Age" in df.columns and "Fare" in df.columns and "Survived" in df.columns:
    fig = px.scatter(df, x="Age", y="Fare", color="Survived",
                     title="Age vs Fare (colored by Survival)")
    fig.show()

print("\n EDA Completed Successfully!")
```
